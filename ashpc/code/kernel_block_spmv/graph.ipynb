```python
import matplotlib.pyplot as plt
import numpy as np 

from dataclasses import dataclass
```

```python
@dataclass
class Resource:
    bram: int
    dsp: int
    ff: int
    lut: int

    @property
    def sum(self) -> int:
        return self.bram + self.dsp + self.ff + self.lut

reports = {
    "32": Resource(2,3,17457,25715),
    "64": Resource(0,3,17472,25723),
    "128": Resource(16,3,17487,25795),
    "256": Resource(18,3,17438,25617),
    "512": Resource(18,3,17453,25679),
}
```

```python
fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(reports.keys(), [r.sum for r in reports.values()])


plt.show()
```
